```python
import pandas as pd
import numpy as np
from ast import literal_eval
import csv
import os as os
```

```python
data = pd.read_csv("dataTest.csv")

processedCSV = 'processed_data.csv'

if not os.path.exists(processedCSV):    
    df = pd.DataFrame(columns=['img', 'joints', 'cartesians', 'yolo_midpoint', 'our_midpoint', 'class'])
    df.to_csv(processedCSV, mode='x', header=True, index=False)
else:
    print(f"{processedCSV} already exists!")
```

```python
# Puts string into a list
res = pd.DataFrame(data["joints"].map(literal_eval).tolist(), columns=['joints'])
res1 = pd.DataFrame(data["cartesians"].map(literal_eval).tolist(), columns=['cartesians'])
res2 = pd.DataFrame(data["yolo midpoints"].map(literal_eval).tolist(), columns=['yolo_midpoints'])
res3 = pd.DataFrame(data["our midpoints"].map(literal_eval).tolist(), columns=['our_midpoints'])

"""
The code below simply grabs the data collected from the 
collect.py script (Joint model data collection)
and cleans it for easier model training. This will be saved
inside of 'processed_data.csv'.

Args:
    None

Returns:
    None
"""
try:
    for i in range(len(data['joints'])):        

        ############ JOINTS ######################
        # Splits based on '_'
        newJoint = res['joints'].str.split('_').tolist()

        # Saves joint and label
        joint = np.array(newJoint, dtype=float)
        joint = joint[i].tolist()
        
        label = joint[6]
        joint = joint[0:6]    

        ############## CARTESIANS #####################
        # Grabs cartesian data column
        newcart = np.array(res1['cartesians']).tolist()

        # Splits data into cartesian and pose
        newcart = newcart[i][2:-6].split('_')    
        test = newcart[2].split(',')
        splitString1 = test[0][:-1]
        splitString2 = test[1][2:len(test[1])]

        # Creates list1 (cartesian) and list2 (pose)
        list1 = [0] * 3
        list1[0:2] = newcart[0:2]
        list1[2] = splitString1

        list2 = [0] 
        list2[0] = splitString2
        list2[1:3] = newcart[3:6]

        # Return data into a list
        list1 = np.array(list1, dtype=float)
        list1 = list1.tolist()

        list2 = np.array(list2, dtype=float)
        list2 = list2.tolist()

        # Creates list of lists that holds both cartesian and pose list
        finalList = [0] * 2
        finalList[0] = list1
        finalList[1] = list2    

        ############## YOLO MIDPOINTS ####################
        # Grabs yolo midpoints column
        newJoint = res2['yolo_midpoints'].str.split('_').tolist()

        # Saves yolo midpoint
        yolo = np.array(newJoint, dtype=float)
        yolo = yolo[i].tolist()

        label = yolo[2]
        yolo = yolo[0:2]        

        ############## OUR MIDPOINTS ###########################
        # Grabs our midpoint column
        newJoint = res3['our_midpoints'].str.split('_').tolist()

        # Saves our midpoint 
        ourMid = np.array(newJoint, dtype=float)
        ourMid = ourMid[i].tolist()
        
        ourMid = ourMid[0:2]        

        # Saves data into a dictionary 
        new_row = {
            'img': data['img'][i],
            'joints': joint,
            'cartesians': finalList,
            'yolo_midpoints': yolo,
            'our_midpoints': ourMid,
            'class': label
        }

        # Inputs data into row of csv file
        new_input = pd.DataFrame([new_row])
        new_input.to_csv(processedCSV, mode='a', header=False, index=False)
    
    print("Completed data processing.")
except:
    print("Something went wrong please try again.")
```
